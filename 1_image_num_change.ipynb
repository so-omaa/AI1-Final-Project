```python
import os
import glob
import cv2
from google.colab.patches import cv2_imshow
```

```python
images = '/content/drive/MyDrive/indian/humans/1 (1).jpg'
image = cv2.imread(images)
cv2_imshow(image)
```

```python
img_path = '/content/drive/MyDrive/indian/humans'  # 폴더 경로
os.chdir(img_path)  # 해당 폴더로 이동
img_names = os.listdir(img_path)  # 해당 폴더에 있는 파일 이름을 리스트 형태로 받음
img_names
```

```python
len(img_names)
```

```python
for name in img_names :  #1 (1).jpg
  if name[-3:] == 'jpg' and name[-5] == ')' :
    src = os.path.join(name)
    print(src)
    dst = src[3:-5] + '.jpg'
    dst = os.path.join(dst)
    print(dst)
    os.rename(src, dst)
```
